```python
import scipy.io as io
import os
import numpy as np
from matplotlib import pyplot as plt
#import pandas as pd
```

```python
file_path = 'Sample_EIT_Data/'
file_name = os.path.join(file_path,"COV077_05.mat")
```

```python
data = io.loadmat(file_name)
```

```python
data.keys()
```

```python
print('The type of the object data is ' + str(type(data['data'])) + ' and its shape is ' + str(data['data'].shape) + '\n')
print('The type of the object section is ' + str(type(data['section'])) + ' and its shape is ' + str(data['section'].shape) + '\n')
```

```python
for i in range(len(data['data'][0][0])):
    print('\nVariable '+str(i)+'\nType: '+str(type(data['data'][0][0][i]))+'\nShape: '+str(data['data'][0][0][i].shape)+'\nSample:\n')
    #print(data['data'][0][0][i])
    print('===================================================================================================')
    # Variables # 43 60 69 72 74 77 79 87 88 contain image data
```

```python
vars = np.array([43, 60, 69, 72, 74, 77, 79, 87, 88])
names = np.array(['lungShape32','TIDCmap','dTIDCmap','dEELmap','ODCLmap','RDVmap','T75map','CMAXmap','PMAXmap'])
for i,var in enumerate(vars):
    if (data['data'][0][0][var].shape[0]==1):
        img = np.reshape(data['data'][0][0][var],(32,32))
        plt.figure()
        plt.imshow(img)
        plt.title('variable '+str(var)+': '+str(names[i]))
        plt.show()
    else:
        plt.figure()
        for j in range(data['data'][0][0][var].shape[0]):
            img = np.reshape(data['data'][0][0][var][j],(32,32))
            plt.subplot(1,data['data'][0][0][var].shape[0],j+1)
            plt.imshow(img)
            if j==0:plt.title('variable '+str(var)+': '+str(names[i]))
        plt.show()
```

```python
for i in range(len(data['section'][0][0])):
    print('\nVariable '+str(i)+'\nType: '+str(type(data['section'][0][0][i]))+'\nShape: '+str(data['section'][0][0][i].shape)+'\nSample:\n')
    #print(data['section'][0][0][i])
    print('===================================================================================================')
# Variables # 13 14 16 47? 48? 58 contain image data
```

```python
vars = np.array([13, 14, 16, 47, 48, 58]) # still missing 47?
names = np.array(['ILungShape32','INoVentPix','xCorrMap','A','dzMov','TidMap'])
for i, var in enumerate(vars):
    if (data['section'][0][0][var].shape[0]==1):
        img = np.reshape(data['section'][0][0][var],(32,32))
        plt.figure()
        plt.imshow(img)
        plt.title('variable '+str(var)+': '+str(names[i]))
        plt.axis('off')
        plt.show()
    elif (data['section'][0][0][var].shape[1]==1024): #for the format of 48
        plt.figure()
        for j in range(data['section'][0][0][var].shape[0]):
            img = np.reshape(data['section'][0][0][var][j],(32,32))
            plt.subplot(5,data['section'][0][0][var].shape[0]/5+1,j+1)
            plt.axis('off')
            plt.imshow(img)
            if j==0:plt.title('variable '+str(var)+': '+str(names[i]))
        plt.show()
    elif (data['section'][0][0][var].shape[0]==1024): #for the format of 47
        temp = np.transpose(data['section'][0][0][var])
        plt.figure()
        
        for j in range(data['section'][0][0][var].shape[1]):
            img = np.reshape(temp[j],(32,32))
            plt.subplot(5,data['section'][0][0][var].shape[1]/5+1,j+1)
            plt.imshow(img)
            plt.axis('off')
            if j==0:plt.title('variable '+str(var)+': '+str(names[i]))
        plt.show()
    
```
